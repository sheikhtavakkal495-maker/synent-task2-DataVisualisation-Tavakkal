```python
#Task Title: Data Visualisation
#Dataset: Iris dataset from Kaggle
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r"C:\Users\DELL\Desktop\Iris.csv")
print(df)
```

```python
# VISUALIZATION 1 : BAR CHART
species_count = df["Species"].value_counts()
plt.figure(figsize=(6,4))
plt.bar(species_count.index, species_count.values)
plt.title("Count of Each Iris Species")
plt.xlabel("Species")
plt.ylabel("Number of Flowers")
plt.show()
```

```python
# VISUALIZATION 2 : HISTOGRAM
plt.figure(figsize=(6,4))
plt.hist(df["SepalLengthCm"], bins=10)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# VISUALIZATION 3 : SCATTER PLOT
plt.figure(figsize=(6,4))
plt.scatter(df["SepalLengthCm"], df["PetalLengthCm"])
plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
# FEATURE COMPARISON
avg_sepal = df.groupby("Species")["SepalLengthCm"].mean()
plt.figure(figsize=(6,4))
plt.bar(avg_sepal.index, avg_sepal.values)
plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")
plt.show()
```
